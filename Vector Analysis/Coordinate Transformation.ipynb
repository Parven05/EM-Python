```python
try:
    import pkg_resources
except ImportError:
    import pip._vendor.pkg_resources as pkg_resources
    import sys
    sys.modules['pkg_resources'] = pkg_resources

from vpython import *

def example():
    x = 3
    y = 4
    z = 5

    # cylindrical
    rho = sqrt(x**2 + y**2)
    phi = atan2(y, x)

    print("Cylindrical")
    print(f'rho = {rho}')
    print(f'phi = {phi}')
    print(f'z = {z}')
    print("--------------------------")

    # Spherical
    r = sqrt(x**2 + y**2 + z**2)
    theta = acos(z / r)
    phi = atan2(y, x)

    print("Spherical")
    print(f'r = {r}')
    print(f'theta = {theta}')
    print(f'phi = {phi}')
    
def cartesian_to_cylindrical_spherical():
    x = 2
    y = 5
    z = 2

    # Cylindrical
    rho = sqrt(x**2 + y**2)
    phi = atan2(y, x)

    print("(i) Cartesian to Cylindrical and Spherical")
    print("<-- Cylindrical -->")
    print(f'rho = {rho}')
    print(f'phi = {phi}')
    print(f'z = {z}')
    print("--------------------------")

    # Spherical
    r = sqrt(x**2 + y**2 + z**2)
    theta = acos(z / r)
    phi = atan2(y, x)

    print("<-- Spherical -->")
    print(f'r = {r}')
    print(f'theta = {theta}')
    print(f'phi = {phi}')
    print("--------------------------")

def cylindrical_to_cartesian():
    rho = 5
    phi = pi / 4
    z = 3

    # Cartesian
    x = rho * cos(phi)
    y = rho * sin(phi)
    z = z

    print("(ii) Cylindrical to Cartesian")
    print(f'x = {x}')
    print(f'y = {y}')
    print(f'z = {z}')
    print("--------------------------")

def spherical_to_cartesian():
    r = 6
    theta = pi / 3
    phi = pi / 4

    x = r * sin(theta) * cos(phi)
    y = r * sin(theta) * sin(phi)
    z = r * cos(theta)

    print("(iii) Spherical to Cartesian")
    print(f'x = {x}')
    print(f'y = {y}')
    print(f'z = {z}')
    print("--------------------------")
    
#example()
cartesian_to_cylindrical_spherical()
cylindrical_to_cartesian()
spherical_to_cartesian()
```
