```python
try:
    import pkg_resources
except ImportError:
    import pip._vendor.pkg_resources as pkg_resources
    import sys
    sys.modules['pkg_resources'] = pkg_resources

from vpython import *

# Scene
WIDTH = 950
HEIGHT = 500
COLOR = vec(0.9, 0.9, 0.9)

# Camera
camera_up = vec(0, 0, 1)
camera_pos = vec(18, 5, 5)
camera_axis = vec(-20, -4, -4)

# colors
red = color.red
green = color.green
blue = color.blue
yellow = color.yellow
black = color.black

# Arrow 
origin = vec(0, 0, 0)
X = vec(10, 0, 0)
Y = vec(0, 10, 0)
Z = vec(0, 0, 10)
width = 0.1

def add_scene():
    scene = canvas(width = WIDTH, height = HEIGHT, background = COLOR)

    scene.up = camera_up
    scene.camera.pos = camera_pos
    scene.camera.axis = camera_axis

    x = arrow(pos = origin, axis = X, shaftwidth = width, color = red)
    y = arrow(pos = origin, axis = Y, shaftwidth = width, color = green)
    z = arrow(pos = origin, axis = Z, shaftwidth = width, color = blue)

    label(pos = x.pos + x.axis, text = "x", color = black)
    label(pos = y.pos + y.axis, text = "y", color = black)
    label(pos = z.pos + z.axis, text = "z", color = black)

def simulate_a():
    P = vec(-2, 0, 3)
    Q = vec(0, 5, 1)
    R = vec(10, 1, -4)

    p = arrow(pos = origin, axis = P, shaftwidth = width, color = yellow)
    q = arrow(pos = origin, axis = Q, shaftwidth = width, color = yellow)
    r = arrow(pos = origin, axis = R, shaftwidth = width, color = yellow)

    label(pos = p.pos + p.axis, text = 'P', color = black)
    label(pos = q.pos + q.axis, text = 'Q', color = black)
    label(pos = r.pos + r.axis, text = 'R', color = black)

    # Calculation
    Out = 2*P + Q - 2*R

    out = arrow(pos = origin, axis = Out, shaftwidth = width, color = yellow)
    label(pos = out.pos + out.axis, text = 'Output', color = black)
    
def simulate_b():
    P = vec(-2, 0, 3)
    Q = vec(0, 5, 1)
    R = vec(10, 1, -4)

    # Calculation
    new_R = 2*R
    Out = dot(P, new_R)
    print("Dot Product :", Out)

def simulate_c():
    x = 10
    y = 1
    z = -4

    # Calculation
    # Cylindrical
    rho = sqrt(x**2 + y**2)
    phi = atan2(y, x)

    print("(i) Cartesian to Cylindrical and Spherical")
    print("<-- Cylindrical -->")
    print(f'rho = {rho}')
    print(f'phi = {phi}')
    print(f'z = {z}')
    print("--------------------------")

    # Spherical
    r = sqrt(x**2 + y**2 + z**2)
    theta = acos(z / r)
    phi = atan2(y, x)

    print("<-- Spherical -->")
    print(f'r = {r}')
    print(f'theta = {theta}')
    print(f'phi = {phi}')
    print("--------------------------")
    
add_scene()
#simulate_a()
#simulate_b()
#simulate_c()
```
