```python
try:
    import pkg_resources
except ImportError:
    import pip._vendor.pkg_resources as pkg_resources
    import sys
    sys.modules['pkg_resources'] = pkg_resources

from vpython import *

# Scene
WIDTH = 950
HEIGHT = 500
COLOR = vec(0.9, 0.9, 0.9)

# Camera
camera_up = vec(0, 0, 1)
camera_pos = vec(18, 5, 5)
camera_axis = vec(-20, -4, -4)

# colors
red = color.red
green = color.green
blue = color.blue
yellow = color.yellow
black = color.black

# Arrow 
origin = vec(0, 0, 0)
X = vec(10, 0, 0)
Y = vec(0, 10, 0)
Z = vec(0, 0, 10)
width = 0.1

def add_scene():
    scene = canvas(width = WIDTH, height = HEIGHT, background = COLOR)

    scene.up = camera_up
    scene.camera.pos = camera_pos
    scene.camera.axis = camera_axis

    x = arrow(pos = origin, axis = X, shaftwidth = width, color = red)
    y = arrow(pos = origin, axis = Y, shaftwidth = width, color = green)
    z = arrow(pos = origin, axis = Z, shaftwidth = width, color = blue)

    label(pos = x.pos + x.axis, text = "x", color = black)
    label(pos = y.pos + y.axis, text = "y", color = black)
    label(pos = z.pos + z.axis, text = "z", color = black)

def vec_magnitude():
    A = vec(3, -1, -2)
    B = vec(-5, 3, 4)

    a = arrow(pos = origin, axis = A, shaftwidth = width, color = yellow)
    b = arrow(pos = origin, axis = B, shaftwidth = width, color = yellow)

    label(pos = a.pos + a.axis, text = 'A', color = black)
    label(pos = b.pos + b.axis, text = 'B', color = black)

    # (i) Magnitude A
    length_A = mag(A)
    label(pos = A * 0.5, text = f"|A| = {A.mag:.2f}", color = black)

    # (ii) Magnitude B
    length_B = mag(B)
    label(pos = B * 0.5, text = f"|B| = {length_A:.2f}", color = black)
    
add_scene()
vec_magnitude()
```
