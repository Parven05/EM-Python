```python
try:
    import pkg_resources
except ImportError:
    import pip._vendor.pkg_resources as pkg_resources
    import sys
    sys.modules['pkg_resources'] = pkg_resources

from vpython import *

def vec_dot():
    A = vec(3, 4, 1)
    B = vec(2, 1, 5)
    C = vec(3, 0, 4)
    D = vec(2, 1, 2)

    dot_AB = A.dot(B)
    dot_CD = C.dot(D)

    magA = A.mag
    magB = B.mag
    magC = C.mag
    magD = D.mag

    theta = diff_angle(A, B)
    theta = acos(dot_AB / (magA * magB))
    angle = degrees(theta)

    print("(i) A.B")
    print("Dot Product : ", dot_AB)
    print("Angle (Degrees) : ", angle)
    print("--------------------------")

    theta = diff_angle(C, D)
    theta = acos(dot_CD / (magC * magD))
    angle = degrees(theta)
    
    print("(ii) C.D")
    print("Dot Product : ", dot_AB)
    print("Angle (Degrees) : ", angle)
    
vec_dot()
```
